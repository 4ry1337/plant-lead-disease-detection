```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

image_size= (256, 256)

image_paths = [
    './dataset/Apple___Cedar_apple_rust/025b2b9a-0ec4-4132-96ac-7f2832d0db4a___FREC_C.Rust 3655.JPG',
    './dataset/Raspberry___healthy/09f1efaf-76ba-419c-a993-2bc1b195a5b4___Mary_HL 9255.JPG',
    './dataset/Tomato___Tomato_Yellow_Leaf_Curl_Virus/032d18ff-d9a6-4e7a-b99a-4bd8e45abe66___YLCV_GCREC 2705.JPG'
]

for image_path in image_paths:
    image = Image.open(image_path)
    original = np.array(image)
    fig, ax = plt.subplots(1, 2, figsize=(12, 5))
    ax[0].imshow(original)
    ax[0].set_title("Original")
    colors = ['red', 'green', 'blue']
    if len(original.shape) == 3:
        for i, color in enumerate(colors):
            ax[1].hist(original[..., i].ravel(), bins=50, color=color, alpha=0.7, label=color.capitalize())
    else:
        ax[1].hist(original.ravel(), bins=50, color='gray', alpha=0.7, label='Gray')
    ax[1].set_title('Pixel Distribution')
    ax[1].set_xlabel('Pixel Value')
    ax[1].set_ylabel('Frequency')
    ax[1].legend()
    plt.tight_layout()
    plt.show()

    image.thumbnail(image_size)
    image_array = np.array(image) / 255.0
    normalized = (image_array * 255).astype(np.uint8)
    fig, ax = plt.subplots(1, 2, figsize=(12, 5))
    ax[0].imshow(normalized)
    ax[0].set_title("Normalized")

    if len(image_array.shape) == 3:  # Check if image is RGB
        for i, color in enumerate(colors):
            ax[1].hist(image_array[..., i].ravel(), bins=50, color=color, alpha=0.7, label=color.capitalize())
    else:
        ax[1].hist(image_array.ravel(), bins=50, color='gray', alpha=0.7, label='Gray')
    ax[1].set_title('Normalized Pixel Distribution')
    ax[1].set_xlabel('Normalized Pixel Value')
    ax[1].set_ylabel('Frequency')
    ax[1].legend()

    plt.tight_layout()
    plt.show()
```
